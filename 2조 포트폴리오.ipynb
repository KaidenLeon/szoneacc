# **01 프로젝트 개요**


## 프로젝트 요약
- 프로젝트명: 서울시 어린이 교통사고 데이터 분석을 통한 어린이 보호구역 개선 방안
- 배경: 공공데이터를 활용하여 어린이 교통사고의 특징과 실태를 살펴보고 개선 방안을 제안하고자 함
- 분석 방법: 서울시 교통사고, 어린이 교통사고 다발지역, 어린이보호구역 관련 데이터 시각화 분석
- 개선 방안: 공공 개선 방안과 민간 개선 방안을 제안
- 구성원 및 역할
  - 구성봉: 교통사고위험지수 산출 / 교통사고위험지수 시각화 / 코로나와 민식이법 상관성 분석 / 스쿨존내 어린이 교통사고의 특징 분석 / 결론도출을 위한 의견취합 및 내용 정리
  - 김현우 
스쿨존사고다발지역 데이터 수집, 전처리 / 교통단속카메라 데이터 수집, 전처리 및 시각화 / 사고다발지역 반경 단속카메라 대수 시각화 및 분석
  - 김성배: 어린이보호구역 데이터 수집, 전처리 / 지도 통합 시각화: 스쿨존, 사고다발지역 / 서울시 어린이 사고다발지역 GIS 시각화 및 분석
  - 홍성희: 보행어린이 사고다발지역 데이터 전처리 / 사고다발지역 및 스쿨존 내 사상자수 시각화 및 분석 / 시간대별 어린이 교통사고 발생건수 시각화 및 분석

## **1.1 프로젝트 배경**
### **1.1.1 어린이 교통사고에 대한 문제 인식**

#### 1) 어린이 교통사고의 정의
어린이 교통사고는 만13세 미만(12세 이하)의 어린이 사상자가 발생한 교통사고입니다.

#### 2) 어린이 교통사고의 원인
- 어린이 교통사고의 원인은 보행자 보호 의무 위반(전방주시 태만, 휴대전화 사용, 운전 미숙 등), 불법 주정차, 신호위반, 안전거리 미확보, 교통안전시설의 유지/관리 미비, 어린이보호구역 지정 대상의 제한, 교통안전교육 부족 등이 있습니다.


#### 3) 민식이법 논란
- 2020년 3월 25일부터 민식이법(특정범죄 가중처벌 등에 관한 법률)이 시행되었습니다. 민식이법은 스쿨존 내에서 발생한 사고의 어린이 보호의무 위반 가해운전자를 가중처벌한다는 내용입니다.
  - 도로교통법 개정안: 어린이보호구역 내 신호등과 과속단속카메라 설치 의무화
  - 특정범죄 가중처벌 등에 관한 법률 개정안: 어린이보호구역 내 안전운전 의무 부주의로 사망이나 상해사고를 일으킨 가해자를 가중처벌
- 민식이법 시행 이후 스쿨존이 많이 신설되었습니다. 서울시의 경우 2018년 27개, 2019년 42개, 2020년 73개가 신설되었습니다.
- 일각에서는 사고 건수는 줄었지만 어린이 교통사고는 지속적으로 발생하고 있으며, 처벌 강화만 할 것이 아닌 다방면의 조치를 취해야 한다는 지적이 나오고 있습니다.
  >-  "이제부터는 지역 맞춤형, 그리고 초등학교 특성에 맞는, 통행 특성에 맞는 어린이보호구역 시설을 강화할 필요가 있습니다."   
  > \- 가천대학교 도시계획학과 교수 장일준 (2021년 3월 23일 YTN 인터뷰)

### **1.1.2 데이터 분석을 통한 개선 방안을 제안**

- 첫 째, 민식이법의 실효성을 분석하고
- 둘 째, 서울시 어린이 사고다발지역의 특성을 분석하고
- 셋 째, 서울시 어린이보호구역의 위험도를 분석하여
- 최종적으로 어린이보호구역의 개선 방안을 보행자와 운전자의 관점에서 제안하고자 합니다.

# **02 분석 방법 및 결과**



## **2.1 데이터 수집**
프로젝트에 사용된 데이터는 아래와 같습니다.
- 서울시 스쿨존 지정 건수 (2016~2020, 서울 열린데이터 광장)
- 보행어린이 사고 다발지역 정보 (2016~2020, TAAS 교통사고 분석시스템)
- 스쿨존 내 어린이 사고다발지역 정보 (2016~2020, TAAS 교통사고 분석시스템)사고건수,사상자수,위치등에 대한 데이터
- 서울시 어린이 인구현황 (2021, 공공데이터 포털)
- 서울시 교통 단속 카메라 (2016~2020, 공공데이터 포털)
- 서울시 전체 교통사고 데이터 (2016~2020, TAAS 교통사고 분석시스템)

- 월별 어린이보호구역내 교통사고 데이터 (2016~2020, TAAS 교통사고 분석시스템)
- 요일별 어린이보호구역내 교통사고 데이터 (2016~2020, TAAS 교통사고 분석시스템)
- 시간대별 어린이보호구역내 교통사고 데이터 (2016~2020, TAAS 교통사고 분석시스템)


## **2.2 데이터 분석: 방법**

### **2.2.1 민식이법의 실효성 분석**
- 민식이법의 실효성을 분석하기 위해 어린이 교통사고 데이터를 시각화하고 시행월(2020년 3월)을 기준으로 분석하였습니다.
- 또한 민식이법이 적용되는 어린이보호구역내 교통사고의 특성 분석을 위해 어린이보호구역내 월별, 요일별, 시간대별 교통사고 데이터를 시각화하였습니다.

### **2.2.2 서울시 어린이 사고다발지역 GIS 분석**
- 어린이 사고다발지역의 특성을 분석하기 위해 최근 5년의 다발지역 데이터를 시각화하고 행정구역별로 분석하였습니다.
- 사고다발지역의 ARI(교통사고위험지수)를 산출하였습니다.
- 추가로 신호/과속 단속카메라, 불법주정차 단속카메라 데이터를 이용하여 사고다발지역 반경 내에 있는 교통안전시설의 개수를 산출하였습니다.

### **2.2.3 서울시 어린이보호구역별 위험도 분석**
- 서울시 어린이보호구역의 위험도를 분석하기 위해 스쿨존 분포도를 지도에 시각화하였습니다.
- 어린이보호구역의 객관적인 위험도를 나타내기 위해 CARI(교차면적 교통사고위험지수)를 산출하였습니다.

### **2.3.2 서울시 어린이 사고다발지역 GIS 분석**
- 어린이 사고다발지역의 정의는 다음과 같습니다.
  - 보행어린이 사고다발지역: 반경 200m에서 보행어린이사고가 3건 이상 또는 사망사고 2건 이상인 지점
  - 스쿨존내어린이 사고다발지역: 어린이 보호구역내 반경 300m에서 어린이 교통사고가 2건 이상 또는 사망사고 1건 이상인 지점
- 어린이 사고다발지역은 어린이 교통사고가 발생하기 쉬운 조건을 가진 지역입니다. 따라서 사고다발지역 데이터를 분석하여 어린이 교통사고의 감소 방안을 찾고자 합니다.

#### 1) 서울시 사고다발지역 분포
- 2016~2020년 동안 지정된 서울시 사고다발지역의 위치와 행정구역별 다발지역 개수, 어린이 인구수를 나타냈습니다.
  - 사고다발지역은 총 105개이며 보행어린이 사고다발지역이 67개, 스쿨존내어린이 사고다발지역이 38개입니다.
  - 사고다발지역이 가장 많은 지역은 강서구 12개, 성북구 9개, 그리고 강남구 강동구가 8개였습니다.
  - 마포구, 중구, 성동구, 광진구는 해당 기간 동안 지정된 다발지역이 없었습니다.
  - 대체로 어린이 인구수가 적은 지역에 다발지역 개수도 적은 경향을 보였습니다.

#### 2) 사고다발지역 ARI
- ARI(Accident Risk Index)는 교통사고 위험지수로 일정 구역의 교통사고 위험도를 나타내는 지표입니다.
- 어린이보호구역의 안전시설을 설치할 때 고려해야할 사항 중 하나로 도로교통법 시행규칙에서 명시하고 있습니다.
  - ARI 계산 방법은 아래와 같습니다.
  > $$ARI = {\sqrt{KSI^2+사고건수^2} \over 년}$$ 
    - KSI(Killed or Seriously Injured: 사고심각성 대표 지표): 사망자 수 + 중상자 수 (어린이 보호구역 내)
    - 사고건수: 인명피해사고 건수
- ARI가 높을수록 위험한 지역으로 판단합니다.
- 사고다발지역 각각의 위험지수를 산출하고 히스토그램으로 나타냈습니다.
  - 3.15의 위험지수가 가장 많았고 수치가 큰 순서대로 5.85과 5.65, 5.4를 기록한 곳은 각각 동작구 중앙대학교병원 부근, 강남구 경남아파트 부근, 구로구 성은어린이집 부근이었습니다.

  ![25](https://github.com/grbluesb/MCPJT/blob/main/ACC/IMG/25.PNG?raw=true)

#### 2) 스쿨존의 위험도 산출: 새로운 지표의 필요성
- 2.3.2에서는 GIS를 통해 사고 다발지역에 대한 분석을 시행하였습니다. 이번에는 사고 다발지역과 어린이보호구역을 함께 분석해보고자 합니다.
- 하지만 현재 보유한 데이터만으로는 서울시 내의 각 스쿨존의 위험도를 판단할만할 기준이 없습니다.
- 사고 다발지역은 사고건수와 사망자, 중상자 데이터를 통해 고유의 교통사고위험지수(ARI)를 산출하였습니다.
- 스쿨존의 경우 근처에서 발생한 각각의 교통사고 데이터가 없기 때문에 위험지수를 산출할 수 없습니다.
  - 사고다발지역 위험도: 반경내 사고건수, 사망자, 중상자 -> ARI
  - 스쿨존 위험도: 반경내 사고건수, 사망자, 중상자 데이터 없음
- 따라서 스쿨존 안에 있는 사고다발지역들의 위험지수(ARI)를 참조하여 각 스쿨존의 인접 위험도를 추측하였습니다.
- 즉 스쿨존 반경 내에 위험지수가 높은 사고다발지역이 많을수록 스쿨존의 위험지수도 함께 높아지게 됩니다.
- 본 프로젝트에서는 어린이보호구역의 위험도를 나타내는 새로운 지표를 CARI(Cross-area ARI, 교차면적 교통사고위험지수)라고 하겠습니다.
- 통계적으로 정확한 산출이 아니며, 다발지역과 인접한 스쿨존의 상대적 위험도 시각화를 위한 단순 산출 방식을 사용했습니다.

  ![35](https://github.com/grbluesb/MCPJT/blob/main/ACC/IMG/35.PNG?raw=true)

- 산출된 CARI를 기반으로 지도상에 스쿨존 분포를 나타냈습니다.

  ![36](https://github.com/grbluesb/MCPJT/blob/main/ACC/IMG/36.PNG?raw=true)

# **03 결론 및 개선 방안**


## **3.1 결론**
1) 민식이법의 실효성
- 2019년 스쿨존내 어린이 사상자는 117명, 민식이법이 시행된 2020년은 사상자 70명으로 대폭 감소했습니다.
- 코로나로 인한 사회적 거리두기와 초등학교 등교 중단 시기와 연관성이 있어 보입니다.
- 민식이법 시행과의 연관성은 코로나 유행 이전과 동일한 상황에서의 비교가 필요할 것으로 생각됩니다.

2) 서울시 어린이 사고다발지역 GIS 분석
- 서울시내 어린이 인구수가 적은 지역은 사고다발지역이 없거나 적은 경향을 보입니다.
- 어린이 교통사고 다발지역 개선이 시급한 행정구역은 강서구와 강동구입니다. 5년 연속 어린이 사고다발지역이 지정된 강서구와 강동구는 다발지역 개선에 예산 투입이 필요하다고 생각됩니다.
- 개선이 시급한 사고다발지역은 강서구 화곡동 신정초교 부근(5년간 4회 지정), 강남구 도곡동 언주초교 부근과 관악구 신림동 애동유치원 부근(5년간 3회 지정)입니다. 두 지역은 지속적으로 사고다발지역이 발생하고 있어 개선이 시급하다고 판단됩니다.
- 사고다발지역의 상당수(87%)가 반경내에 신호/과속카메라가 설치되어 있지 않았고, 불법주정차 단속카메라의 경우 다발지역의 약 42%에 설치되어 있지 않았습니다. 개선을 위해서는 단속카메라 설피가 필요할 것으로 보입니다.

3) 서울시 어린이보호구역의 위험도
- 어린이 교통사고가 많이 발생하는 5~6월, 평일, 하교시간을 고려한 맞춤 어린이 보호 정책이 필요할 것으로 생각됩니다.
- 서울시내 사고다발지역과 인접한 394개 스쿨존의 교통 안전 환경을 우선적으로 개선하는 것이 효율적이라고 판단됩니다.

- 기존에는 전방에 어린이보호구역을 안내하고 있지만 어린이 안전 옵션을 켜면 추가적인 안내가 포함되는 방식입니다.
- 만약 운전자가 어린이 교통사고 위험이 높은 하교 시간대에 스쿨존에 진입한다면 어린이 사고 주의 시간대라는 것을 추가적으로 안내하는 것입니다.
- 또한 평일인지 공휴일인지에 따라서도 주의 내용을 안내한다면 조금이나마 어린이 교통사고를 줄일 수 있을 것이라고 생각됩니다.

## **3.3 한계점**
#### 1) 다발지역 이외의 교통사고 위치 정보가 고려되지 않음
- 서울시 어린이 교통사고 다발지역 외에서 발생한 교통사고 데이터의 부재로 다발지역의 데이터만 가지고 추측성 분석을 하였습니다.
- 각 스쿨존의 위험도를 분석할 때 스쿨존에서 일어난 실제 사고가 아닌, 스쿨존 인근의 다발지역 데이터를 참조하였습니다.
- 스쿨존 반경 내에서 사고가 발생했다고 하더라도 그 지역이 다발지역으로 지정되지 않았다면 위험도가 0으로 산정됩니다.
- 따라서 서울시 스쿨존의 위험도 순위는 다발지역 인근에 위치한 스쿨존에 한해서 상대적인 위험도 순위로만 활용해야 합니다.

#### 2) 어린이보호구역 지정 시점과 사고다발지역 지정 시점이 고려되지 않음
- 어린이보호구역과 사고다발지역은 2016년부터 2020년까지 5년 사이의 데이터를 일괄적으로 분석하였습니다.
- 사고다발지역의 경우, 사고다발지역으로 지정이 된 이후 이미 개선 조치가 이루어졌을 가능성이 있습니다.

#### 3) 향후 개선 사항
- 프로젝트를 위해 수집한 교통안전시설 데이터는 신호/과속단속카메라, 불법주정차단속카메라 이외에도 과속방지턱, 교차로 데이터 등이 있었습니다. 하지만 결측치, 불완전한 위치정보 등의 데이터상의 문제로 본 프로젝트에서는 사용하지 못했습니다.
- 또한 개별적인 교통사고 위치 데이터의 부재로 부득이하게 다발지역의 데이터를 활용하여 스쿨존의 상대적 위험도 순위를 산출하였습니다.
- 앞으로 공공기관에서 공공데이터를 보다 체계적으로 수집, 보완하고 관리한다면 더욱더 가치있는 결과를 만들 수 있을 것으로 생각됩니다.

# **04 개발후기 및 느낀점**
- 홍성희: 프로젝트 주제를 정할 때 욕심은 많은데 짧은 기간 내에 수집할 수 있는 프로젝트가 한정적 이여서 아쉬웠습니다. 그래서 주제를 정하고 데이터 전처리를 하면서 좋은 데이터, 내가 쓸려는 필요한 데이터를 잘 찾고 전처리를 잘 하는 것이 시각화에서 가장 중요하다고 생각하게 되었습니다. 이번 프로젝트를 하면서 프로젝트의 방향을 잃어버려 만족스러운 결론이 나지 않아 힘들었었습니다. 하지만 중간 점검 때 강사님의 피드백을 듣고 친절하지 않은 시각화를 하고 있었다는 생각을 하게 되었습니다. 그래서 팀원들과 오래 대화를 하게 되었고 다시 하나하나 정리하고 새로운 데이터도 가져와서 전과 다른 방향으로 분석을 진행하였습니다. 다시 시작한다는 것이 조금 부담되었지만 프로젝트를 하며 쌓인 기반 지식과 시각화 경험 덕분에 전보다 훨씬 빠르게 원하는 방향으로 프로젝트가 끝맺음을 맺을 수 있게 되어 좋았습니다.
- 김성배: 처음 프로젝트 주제 및 방향을 정하는 부분에서 어려움이 있었습니다. 어떤 문제에 대해서 어떤식으로 접근하여 어떤 결론을 도출할지를 처음에 정하는게 중요하다는 생각이 들었습니다. 그러나 팀원들과 함께 프로젝트 중간에 다시 문제에 대한 접근 방식과 결론을 논의하고, 프로젝트를 진행하여 성공적으로 마무리 할 수 있어 좋았습니다.
- 김현우: 초기 프로젝트 주제 선정을 위해 어려웠던 점은 데이터 검증이었으며 다소 시간이 소비되었으나 개인별 주제 아이디어 발표 및 다수결에 의해 주제 선정에 확정이 이루어졌습니다. 데이터 정제 과정 중 Polygon 좌표 활용을 위해 GeoJSON, GeoPandas에 대해 공부하였고 조원분께 관련 연구 내용을 공유하고 본인의 경우 관련 지식이 부족하여 다른 조원분이 끝까지 연구해서 시각화 결과물이 도출되어 만족된 결과를 만들어냈고 프로젝트 중기에 일부 데이터 프레임 상의 결측치로 인해 채워 넣는 과정이 조금 시간 소비되었고 시각화 테스트 중 Raw data 초기부터 다시 재전 처리하는 과정이 많이 발생하였고 차후 중요해 보이지 않는 데이터라도 누락시키지 않고 진행하여 초기 실수에 빠르게 대응할 수 있게 되었습니다. 시각화 막바지 부분에서 Folium 기술적 문제가 발생하여 다소 시간이 많이 소비되었고 (간헐적으로 마킹이 1000개 이상일 경우 런타임 끊김 발생) 조원분들과 여러 차례 테스트하였으나 한 분만 정상적 작동하였고 이외 조원들은 작동하지 않아 FT 선생님들 도움을 받아 자바스크립트를 이용한 대량 마커 찍기의 기술적 문제를 해결하였습니다. 결론적으로 Folium이 완벽한 툴은 아니었다 생각이 들고 에러 없이도 문제가 발생할 수 있다는 걸 인지하게되었습니다.
- 구성봉: 처음에는 데이터를 보고 프로젝트를 어떤 방향으로 진행해야할지 판단하기가 어려웠습니다. 나름의 판단으로 프로젝트의 방향을 잡고 진행했지만 만족스러운 결과가 나오지 않앗습니다. 하지만 포기하지 않고 다시 처음부터 데이터를 보면서 프로젝트 완수가 아닌 데이터를 해석하는 데에 초점을 맞췄습니다. 또한 팀원들과 의견 조율이 잘 되지 않을때, 지금까지 해왔던 작업들을 뒤로 해야 했을 때 팀원들에게 미안한 마음이 컸습니다. 하지만 어려웠던 부분은 시행착오였으며 결과적으로 데이터를 보는 시각을 키울 수 있던 값진 경험이었습니다.

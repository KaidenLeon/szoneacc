```python
data = pd.read_csv('/content/drive/MyDrive/[D9&10] 데이터시각화 프로젝트_2조/Data/raw data/202112_202112_연령별인구현황_연간.csv', encoding='cp949')
```

```python
data = data.rename(columns={'2021년_계_총인구수': '2021년 총 인구수',
                            '2021년_계_연령구간인구수': '2021년 어린이 인구수',
                            '2021년_계_0세': '0세',
                            '2021년_계_1세': '1세',
                            '2021년_계_2세': '2세',
                            '2021년_계_3세': '3세',
                            '2021년_계_4세': '4세',
                            '2021년_계_5세': '5세',
                            '2021년_계_6세': '6세',
                            '2021년_계_7세': '7세',
                            '2021년_계_8세': '8세',
                            '2021년_계_9세': '9세',
                            '2021년_계_10세': '10세',
                            '2021년_계_11세': '11세',
                            '2021년_계_12세': '12세',
                            '2021년_계_13세': '13세',
                            })
```

```python
data['행정구역'] = data['행정구역'].str.strip('서울특별시').str.strip()
data['행정구역'] = data['행정구역'].str[:-13].str.strip()
data['행정구역'][0] = '서울특별시'
```

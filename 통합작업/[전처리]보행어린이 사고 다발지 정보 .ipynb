```python
# 사고다발지ID를 연도로 변환하기
dataCW['사고다발지ID'] = dataCW['사고다발지ID'].apply(str)
dataCW['사고다발지ID'] = pd.to_numeric(dataCW['사고다발지ID'].str[:-3])
dataCW['사고다발지ID'] = pd.to_numeric(dataCW['사고다발지ID'])
dataCW.head(3)
```

```python
# 사고다발지ID 컬럼명을 연도로 변환하기
dataCW = dataCW.rename(columns={'사고다발지ID':'연도'})
dataCW.head()
```

```python
# 지역이 서울인 데이터 중, 사용할 컬럼(법정동코드', '지점코드', '시도시군구명', '지점명', '위도', '경도', '다발지역폴리곤)만 골라서 새로운 DataFrame 생성
CWA = dataCW[dataCW['지역']=='서울']
col = ['법정동코드', '지점코드', '연도', '행정구역', '지점명','발생건수',	'사상자수',	'사망자수',	'중상자수',	'경상자수',	'부상신고자수', '위도', '경도', '다발지역폴리곤']
CWA = dataCW[dataCW['지역']=='서울'][col]
CWA
```

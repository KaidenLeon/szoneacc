```python
from folium.plugins import FastMarkerCluster

map = folium.Map(location=[lat, lon], tiles='cartodbpositron', zoom_start=11)

callback = ('function (row) {' 
                'var circle = L.circle(new L.LatLng(row[0], row[1]), {color: "red",  radius: 5});'
                'return circle};')
callback = ('function (row) {' 
                'var circle = L.circle(new L.LatLng(row[0], row[1]), {color: "red",  radius: 1000});'
                'return circle};')
callback2 = ('function (row) {' 
                'var circle = L.circle(new L.LatLng(row[0], row[1]), {color: "green",  radius: 5});' #단속카메라 위치
                'return circle};')
callback3 = ('function (row) {' 
                'var circle = L.circle(new L.LatLng(row[0], row[1]), {color: "blue",  radius: 5});'
                'return circle};')
callback3 = ('function (row) {' 
                'var circle = L.circle(new L.LatLng(row[0], row[1]), {color: "blue",  radius: 1000});'
                'return circle};')

map.add_child(FastMarkerCluster(szca[['위도', '경도']].values.tolist(), callback=callback))
map.add_child(FastMarkerCluster(sc[['위도', '경도']].values.tolist(), callback=callback2))
map.add_child(FastMarkerCluster(sza[['위도', '경도']].values.tolist(), callback=callback3))

map
```
